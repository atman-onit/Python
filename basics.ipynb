##Everything in python is an object

```python
# we can use type() function to find class of each object

a = 5
b = 8.9
c = 'h'
d = "helloworld"
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

we can see how data types like int float str are considered as classes and their instances/values in codes are considered as objects.

even a class is an object 
this is fundamental of python called duck typing
