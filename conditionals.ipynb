we can simplfy this code- using != operator


```python
num1 = int(input("enter number 1: "))
print(f"num1 = {num1}")
num2 = int(input("Enter number 2:"))
print(f"num2 = {num2}")

if num1 != num2:
    print(f"num1 not equal to num 2")

    if num1>num2:
        print(f"{num1} > {num2}")

    else:
        print(f"{num1} < {num2}")

else:
    print("Equal numbers")
```

# MATCH CASE
---
we use match case for specific values which when matched executes its block of code,

```python
year = int(input("enter your college year"))

match year:
    case 1:
        print("You're a fresher! ")
    case 2:
        print("You're in second year")
    case 3:
        print("You're in 3rd year ")
    case 4:
        print("You're in 4th year ")
    case _: #if no cases match
        print("invalid")
```
