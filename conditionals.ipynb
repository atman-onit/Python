We use ' or ' keyword to reduce cdde complexity and length

```python
num1 = int(input("enter number 1: "))
print(f"num1 {num1}")
num2 = int(input("Enter number 2:"))
print(f"num2 = {num2}")

if num1 > num2 or num1 < num2:
    print("Unequal numbers")
    if num1>num2:
        print(f"{num1} > {num2}")
    else:
        print(f"{num1} < {num2}")
else:
    print("Equal numbers")
```

we can simplfy this code- using != operator


```python
num1 = int(input("enter number 1: "))
print(f"num1 {num1}")
num2 = int(input("Enter number 2:"))
print(f"num2 = {num2}")

if num1 != num2:
    print(f"num1 not equal to num 2")
    if num1>num2:
        print(f"{num1} > {num2}")
    else:
        print(f"{num1} < {num2}")
else:
    print("Equal numbers")
```
